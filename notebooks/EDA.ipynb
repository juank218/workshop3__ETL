```python
happiness_df = {
    "2015": data_2015,
    "2016": data_2016,
    "2017": data_2017,
    "2018": data_2018,
    "2019": data_2019
}
```

```python
# Find the common columns across all DataFrames
shared_columns = list(set.intersection(*(set(data.columns) for data in year_happiness_df.values())))

# Filter each DataFrame to keep only the common columns
filtered_df = [data[shared_columns] for data in year_happiness_df.values()]

# Concatenate the filtered DataFrames into one, ignoring the original index
merged_df = pd.concat(filtered_df, ignore_index=True)

# Display the first few rows of the resulting DataFrame
merged_df.head()
```

```python
# Fill missing values in the 'corruption_perception' column with the column's mean
merged_df["corruption_perception"] = (
    merged_df["corruption_perception"]
    .fillna(merged_df["corruption_perception"].mean())
)

# Query the DataFrame for rows where the country is 'United Arab Emirates' and the year is '2018'
merged_df.query("country == 'United Arab Emirates' & year == '2018'")
```

```python
# Initialize the CountryConverter
cc = coco.CountryConverter()

# Convert the 'country' column to continents and store the result in a new 'continent' column
merged_df["continent"] = cc.convert(names=merged_df["country"].tolist(), to="continent")

# Display the first 10 rows of the 'continent' and 'country' columns
merged_df[["continent", "country"]].head(10)
```

```python
# Remove duplicate countries to get unique entries
unique_countries = merged_df.drop_duplicates(subset=['country'])

# Get the count of unique countries per continent
unique_countries['continent'].value_counts()
```

```python
# Query for unique countries in the continent "America"
unique_countries.query("continent == 'America'")["country"].unique()
```

```python
# Define a mapping for specific countries to continents
continent_mapping = {
    "Canada": "North America",
    "Costa Rica": "Central America",
    "Mexico": "North America",
    "United States": "North America",
    "Brazil": "South America",
    "Venezuela": "South America",
    "Panama": "Central America",
    "Chile": "South America",
    "Argentina": "South America",
    "Uruguay": "South America",
    "Colombia": "South America",
    "Suriname": "South America",
    "Trinidad and Tobago": "South America",
    "El Salvador": "Central America",
    "Guatemala": "Central America",
    "Ecuador": "South America",
    "Bolivia": "South America",
    "Paraguay": "South America",
    "Nicaragua": "Central America",
    "Peru": "South America",
    "Jamaica": "Central America",
    "Dominican Republic": "Central America",
    "Honduras": "Central America",
    "Haiti": "Central America",
    "Puerto Rico": "Central America",
    "Belize": "Central America",
    "Trinidad & Tobago": "South America"
}
```

```python
# Map the continent information based on the country and fill missing values with existing continent data
merged_df["continent"] = merged_df["country"].map(continent_mapping).fillna(merged_df["continent"])

# Display the first 10 rows of the 'continent' and 'country' columns
merged_df[["continent", "country"]].head(10)
```

```python
# Define the new column order
order = [
    'country',
    'continent',
    'year',
    'health',
    'social_support',
    'economy',
    'corruption_perception',
    'freedom',
    'generosity',
    'happiness_rank',
    'happiness_score'
]

# Reorder the columns in merged_df
merged_df = merged_df[order]

# Display the first few rows of the reordered DataFrame
merged_df.head()
```

```python
# Set the renderer to open in a web browser
pio.renderers.default = "browser"

# Simple choropleth map showing happiness score per country
fig = px.choropleth(
    merged_df,  # Replace with your DataFrame name if different
    locations="country",
    locationmode="country names",
    color="happiness_score",
    hover_name="country",
    color_continuous_scale="Viridis",
    title="Happiness Score per Country"
)

# Display the map in the browser
fig.show()
```

# Merge all DataFrames, Keeping Only Common Columns
We concatenate the five DataFrames within year_happiness_df, keeping only the columns that are shared among all of them.

```python
# Find the common columns across all DataFrames
shared_columns = list(set.intersection(*(set(data.columns) for data in year_happiness_df.values())))

# Filter each DataFrame to keep only the common columns
filtered_df = [data[shared_columns] for data in year_happiness_df.values()]

# Concatenate the filtered DataFrames into one, ignoring the original index
merged_df = pd.concat(filtered_df, ignore_index=True)

# Display the first few rows of the resulting DataFrame
merged_df.head()
```

# Analysis of Average Happiness Scores by Country and Continent
Chart 1: Top 10 Countries by Average Happiness Score

This chart shows the top 10 countries with the highest average happiness scores, with Denmark, Norway, and Finland leading. Most of these countries are in Europe, suggesting that factors like high income, strong social support, and healthcare access may contribute to higher happiness levels.

Chart 2: Average Happiness Score by Continent

This chart illustrates the average happiness scores by continent. Oceania, North America, and Europe have the highest scores, while Africa has the lowest. The regional differences highlight how economic stability and social infrastructure may impact overall happiness.

```python
# Calculate the top 10 countries and continents by average happiness score
top_countries_happiness = merged_df.groupby("country")["happiness_score"].mean().nlargest(10)
top_continents_happiness = merged_df.groupby("continent")["happiness_score"].mean().nlargest(10)

# Create subplots with adjusted size
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot for top 10 countries
top_countries_happiness.plot(kind="bar", ax=axes[0], color="purple")
axes[0].set_title("Top 10 Countries by Average Happiness Score")
axes[0].set_xlabel("Country")
axes[0].set_ylabel("Avg Happiness Score")

# Plot for top 10 continents
top_continents_happiness.plot(kind="bar", ax=axes[1], color="yellow")
axes[1].set_title("Top 10 Continents by Average Happiness Score")
axes[1].set_xlabel("Continent")
axes[1].set_ylabel("Avg Happiness Score")

# Improve layout and display the plots
plt.tight_layout()
plt.show()
```

Chart 3: Bottom 10 Countries by Average Happiness Score

This chart shows the 10 countries with the lowest average happiness scores. The countries include Burundi, Central African Republic, Syria, South Sudan, Rwanda, Tanzania, Afghanistan, Togo, Yemen, and Madagascar. Many of these countries are located in regions experiencing political instability, economic hardship, and limited access to resources. This suggests that factors like conflict, poverty, and lack of infrastructure may contribute significantly to lower happiness levels.


```python
# Calculate the bottom 10 countries by average happiness score
bottom_countries_happiness = merged_df.groupby("country")["happiness_score"].mean().nsmallest(10)

# Create a single plot for bottom 10 countries with adjusted size
plt.figure(figsize=(10, 6))

# Plot for bottom 10 countries
bottom_countries_happiness.plot(kind="bar", color="purple")
plt.title("Bottom 10 Countries by Average Happiness Score")
plt.xlabel("Country")
plt.ylabel("Avg Happiness Score")

# Display the plot
plt.tight_layout()
plt.show()
```

# Correlation Analysis
This correlation matrix displays the relationships between various factors and happiness metrics. The color intensity indicates the strength and direction of the correlation, where values close to 1 or -1 signify strong positive or negative relationships, respectively.

Happiness Score and Economy (0.79): There is a strong positive correlation between the happiness score and the economy, suggesting that countries with higher GDP per capita tend to have higher happiness levels.

Happiness Score and Health (0.74): Health, likely measured by life expectancy or related indicators, is also highly correlated with happiness, indicating that better health outcomes are associated with higher happiness.

Social Support and Happiness Score (0.65): Social support shows a moderate positive correlation with happiness, suggesting that communities with stronger social bonds tend to report higher happiness.

Freedom and Happiness Score (0.55): Freedom exhibits a moderate positive correlation with happiness, implying that personal freedoms contribute to overall happiness.
Happiness Rank and Happiness Score (-0.99): As expected, there is a very strong negative correlation between happiness rank and happiness score, given that a higher score corresponds to a lower rank (better ranking).

Weak Correlations: Generosity and corruption perception have weaker correlations with happiness, indicating that these factors may have a less direct influence on overall happiness compared to economic and health factors.

```python
# Drop non-numeric columns for correlation analysis
numeric_df = merged_df.drop(columns=["country", "continent"])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Set up the figure and plot the heatmap
plt.figure(figsize=(12, 7))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```
